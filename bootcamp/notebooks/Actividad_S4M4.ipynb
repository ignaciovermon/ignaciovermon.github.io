```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl

np.random.seed(42)
temperatura = 20 + 5 * np.random.randn(100)  # Temperatura promedio 20, desviación estándar 5
a = 2
b = 50
ruido = 10 * np.random.randn(100)
consumo_energia = a * temperatura + b + ruido

lista_temperatura = temperatura.tolist()
lista_consumo_energia = consumo_energia.tolist()

df = pd.DataFrame({'Temperatura': lista_temperatura, 'Consumo_Energia': lista_consumo_energia})
df.head()
```

```python
from sklearn.linear_model import LinearRegression

# Preparar los datos para scikit-learn
X = df[['Temperatura']]  # La característica debe ser un DataFrame
y = df['Consumo_Energia']  # La variable objetivo puede ser una Serie

# Crear y ajustar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener e imprimir los coeficientes
intercepto = modelo_regresion.intercept_
pendiente = modelo_regresion.coef_[0]

print(f"Intercepto: {intercepto}")
print(f"Pendiente: {pendiente}")
```

```python
df['Consumo_Energia_pred'] = modelo_regresion.predict(X)
df['error'] = df['Consumo_Energia'] - df['Consumo_Energia_pred']
df.head()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Calcular el Error Cuadrático Medio (MSE)
mse = mean_squared_error(df['Consumo_Energia'], df['Consumo_Energia_pred'])
print(f"Error Cuadrático Medio (MSE): {mse:.2f}")

# Calcular el Error Absoluto Medio (MAE)
mae = mean_absolute_error(df['Consumo_Energia'], df['Consumo_Energia_pred'])
print(f"Error Absoluto Medio (MAE): {mae:.2f}")

# Interpretación de los resultados:
print("\nInterpretación del ajuste del modelo:")
print("El MSE y el MAE miden el promedio de los errores entre los valores predichos y los reales.")
print(f"- Un MSE de {mse:.2f} indica la magnitud de los errores cuadráticos promedio. Como los errores se elevan al cuadrado, este métrica penaliza más los errores grandes.")
print(f"- Un MAE de {mae:.2f} representa la magnitud promedio de los errores absolutos.")
print("En general, valores más bajos de MSE y MAE indican un mejor ajuste del modelo a los datos.")
print(f"un MAE de {mae:.2f} significa que, en promedio, las predicciones del modelo se desvían aproximadamente {mae:.2f} unidades de los valores reales de Consumo de Energía.")
print("El ajuste parece razonable, pero siempre es útil comparar estos valores con otros modelos o con un modelo base simple.")
```
