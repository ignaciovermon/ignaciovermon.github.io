```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {'Edad': np.random.randint(20, 60, 100),
        'Ingresos': np.random.randint(30000, 100000, 100),
        'Años de Educación': np.random.randint(12, 20, 100),
        'Horas de Sueño': np.random.uniform(5.0, 9.0, 100)}

df = pd.DataFrame(data)

df.head()
```

```python
df_numeric = df.select_dtypes(include=np.number)

correlation_matrix = df_numeric.corr()

correlation_matrix
```

```python
plt.figure(figsize=(8, 6))  # Ajustar el tamaño de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5) # Añadir líneas de borde entre las celdas
plt.title('Matriz de Correlación')
plt.show()
```

La correlacion va de -1 a 1, mientras mas cercano a 1 mas correlacionado positivamente. Si esta mas cercano a -1 mas correlacionado negativamente. Si es cercano a 0 no hay correlacion. Por lo que podemos ver una baja correlacion en toda la combinacion de datos.
