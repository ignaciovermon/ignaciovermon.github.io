```python
def calcular_area_rectangulo():
    base = float(input("Ingrese el largo del rectángulo: "))
    altura = float(input("Ingrese el ancho del rectángulo: "))
    area = base * altura
    print(f"El área del rectángulo es: {area}")

calcular_area_rectangulo()
```

```python
import numpy as np
def calcular_circunferencia():
    radio = float(input("Ingrese el radio del círculo: "))
    circunferencia = 2 * np.pi * radio
    print(f"La circunferencia del círculo es: {circunferencia}")
calcular_circunferencia()
```

```python
def calcular_promedio():
  numero_calificaciones = int(input("Ingrese el numero de calificaciones: "))
  suma = 0
  for i in range(numero_calificaciones):
    calificacion = float(input(f"Ingrese la calificacion {i+1}: "))
    suma += calificacion
  promedio = suma/numero_calificaciones
  print(f'El promedio de las calificaciones es: {promedio}')
calcular_promedio()
```

```python
import statistics
def calcular_promedio():
  numero_calificaciones = int(input("Ingrese el numero de calificaciones: "))
  calificaciones = []
  for i in range(numero_calificaciones):
    calificacion = float(input(f"Ingrese la calificacion {i+1}: "))
    calificaciones.append(calificacion)
  promedio = statistics.mean(calificaciones)
  print(f'El promedio de las calificaciones es: {promedio}')
calcular_promedio()
```

```python
import random
def generar_numero_aleatorio():
  cantidad_numeros = int(input("Ingrese la cantidad de numeros aleatorios a generar: "))
  limite_numeros = int(input("Ingrese el limite del numero a generar: "))
  numeros_aleatorios = []
  for _ in range(cantidad_numeros):
    numero_aleatorio = random.randint(1, limite_numeros)
    numeros_aleatorios.append(numero_aleatorio)
  print(f'Los números aleatorios generados son: {numeros_aleatorios}')
generar_numero_aleatorio()
```

```python
def llamar_funciones():
  calcular_area_rectangulo()
  calcular_circunferencia()
  calcular_promedio()
  generar_numero_aleatorio()
llamar_funciones()
```
