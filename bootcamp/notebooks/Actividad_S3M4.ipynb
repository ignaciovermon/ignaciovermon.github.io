```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42) # para reproducibilidad
n_samples = 100
horas_ejercicio = np.random.randint(1, 10, n_samples)
# Simular una relación inversa con ruido
presion_arterial = 120 - (horas_ejercicio * 3) + np.random.normal(0, 5, n_samples)
grupos_edad = np.random.choice(['Joven', 'Adulto', 'Mayor'], n_samples)
tipos_dieta = np.random.choice(['Mediterránea', 'Cetogénica', 'Vegetariana', 'Normal'], n_samples)

df = pd.DataFrame({'Horas Ejercicio': horas_ejercicio, 'Presion Arterial': presion_arterial, 'Grupo Edad': grupos_edad, 'Tipo Dieta':tipos_dieta})
df.head()
```

```python
df_pivot = df.pivot_table(values='Presion Arterial', index='Tipo Dieta', columns='Grupo Edad', aggfunc='mean')
display(df_pivot)
```

```python
sns.scatterplot(data=df, x='Horas Ejercicio', y='Presion Arterial', hue='Grupo Edad')
```

```python
correlacion_pearson = df['Horas Ejercicio'].corr(df['Presion Arterial'],method='pearson')
print('La correlación de Pearson es:', correlacion_pearson, ' en este caso, esta comprobado de que a mayor cantidad de actividad fisica, la presión arterial disminuye, por lo que la correlacion encontrada implica causalidad')
```
