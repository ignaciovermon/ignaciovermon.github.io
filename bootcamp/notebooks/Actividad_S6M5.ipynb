```python
import numpy as np
from scipy.stats import t
```

*   H0: El numero de libros prestado por usuario es 5
*   H1: El numero de libros prestado por usuario es distinto de 5

```python
alpha = 0.05
n = 30
X = 5 #Promedio historico
Libros_prestados = [6, 4, 5, 7, 3, 6, 8, 4, 5, 6, 7, 5, 4, 6, 5, 7, 6, 4, 5, 8, 5, 6, 4, 7, 5, 6, 4, 5, 7, 6]

# Paso 1: Calcular el estadístico t
grados_libertad = n - 1
media_muestral = np.mean(Libros_prestados)
SE = np.std(Libros_prestados) / np.sqrt(n)
t_estadistico = (media_muestral - X) / SE

# Paso 2: Calcular el valor-p (prueba bilateral)
p_valor = 2 * (1 - t.cdf(abs(t_estadistico), df=grados_libertad))

# Resultados
print(f"Estadístico t: {t_estadistico:.2f}")
print(f"Valor-p: {p_valor:.4f}")

# Decisión
if p_valor < alpha:
    print("Se rechaza H₀. La media es significativamente diferente de 5.")
else:
    print("No se rechaza H₀. No hay suficiente evidencia para concluir que la media es diferente de 5.")
```
