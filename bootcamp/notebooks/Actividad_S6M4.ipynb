```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datos_trafico.csv')
df.head()
```

```python
x = df['Vehículos']
y = df['Día']
plt.figure(figsize=(10,6))
plt.plot(x,y, label = 'Vehiculos por día',color='tab:blue')
plt.title('Vehiculos por día')
plt.legend()
plt.xlabel('Vehículos')
plt.ylabel('Día')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Velocidad_Promedio'],bins=8,color='tab:red',label='Distribucion')
plt.xlabel('Velocidad promedio')
plt.ylabel('Frecuencia')
plt.legend()
plt.title('Velocidad promedio por vehículos')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Vehículos'],df['Accidentes'],color = 'tab:green',label='Accidentes por vehiculos')
plt.xlabel('Vehículos')
plt.ylabel('Accidentes')
plt.legend()
plt.title('Accidentes por vehículos')
plt.savefig('grafico_accidentes.png')
plt.show()
```
