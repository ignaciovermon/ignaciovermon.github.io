```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
data = {'Edad': [25, 30, 35, 40, 45],
        'Ingresos': [50000, 60000, 70000, 80000, 90000],
        'Años de Educación': [16, 18, 17, 19, 20],
        'Horas de Sueño': [7, 6, 8, 7, 6]}

df = pd.DataFrame(data)
display(df)
```

```python
correlation_matrix = df.corr()
display(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor de la Matriz de Correlación')
plt.show()
```

Observando el mapa de calor:

Correlaciones Positivas Fuertes: Los valores cercanos a 1 (colores cálidos) indican una fuerte correlación positiva. En este caso, vemos una correlación positiva fuerte entre Edad e Ingresos, lo que sugiere que a medida que aumenta la edad, también tienden a aumentar los ingresos. También hay una correlación positiva fuerte entre Edad y Años de Educación, y entre Ingresos y Años de Educación.
Correlaciones Negativas: Los valores cercanos a -1 (colores fríos) indican una correlación negativa. Vemos una correlación negativa entre Años de Educación y Horas de Sueño, lo que podría sugerir que las personas con más años de educación tienden a dormir menos horas. También hay una correlación negativa débil entre Edad y Horas de Sueño, y entre Ingresos y Horas de Sueño.
Correlación Nula o Débil: Los valores cercanos a 0 (colores más neutros) indican una correlación débil o nula.
