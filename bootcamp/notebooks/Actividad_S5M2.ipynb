```python
productos = ["Balón de fútbol", "Raqueta de tenis", "Zapatillas de running", "Cuerda de saltar", "Mancuernas"]
```

```python
productos.append("Zapatos de futbol")
productos.append("Canilleras")
productos
```

```python
stock = [25, 10, 30, 50, 15, 20, 40]

inventario = dict(zip(productos, stock))

inventario
```

```python
inventario["Pelota de tenis"] = 20
print(f'El inventario de pelota de tenis es {inventario["Pelota de tenis"]}')
```

```python
productos_unicos = set(productos)

if len(productos_unicos) == len(productos):
  print("No hay elementos duplicados en la lista de productos.")
else:
  print("Existen elementos duplicados en la lista de productos.")

productos_unicos
```
