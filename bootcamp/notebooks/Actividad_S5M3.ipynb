```python
import pandas as pd

datos = {
    "Estudiante": ["Juan", "Juan", "María", "María"],
    "Materia": ["Matemáticas", "Historia", "Matemáticas", "Historia"],
    "Calificación": [6.5, 5.8, 4.2, 6.0]
}

df = pd.DataFrame(datos)
print(df)
```

```python
df_indexed = df.set_index(['Estudiante', 'Materia'])
df_indexed
```

```python
maria_historia = df_indexed.loc[('María', 'Historia'), 'Calificación']
print(f"La clasificación de María en Historia es: {maria_historia}")
```

```python
df_agrupado_materia = df.groupby('Materia')['Calificación'].agg('mean')
df_agrupado_materia
```

```python
df_agrupado_materia_max = df.groupby('Materia')['Calificación'].agg('max')
df_agrupado_materia_max
```

```python
df_pivot = df.pivot(index='Estudiante', columns='Materia', values='Calificación')
df_pivot
```

```python
df_melted = df_pivot.melt(ignore_index=False, var_name='Materia', value_name='Calificación').reset_index()

df_melted.rename(columns={'index': 'Estudiante'}, inplace=True)

print("\nDataFrame despivoteado (melt):")
df_melted
```

```python
df1 = pd.DataFrame({
    "ID_Estudiante": [101, 102, 103, 104],
    "Estudiante": ["Alice", "Bob", "Charlie", "David"],
    "Carrera": ["Ingeniería", "Ciencias", "Artes", "Ingeniería"]
})

df2 = pd.DataFrame({
    "ID_Estudiante": [101, 102, 101, 103],
    "Materia": ["Cálculo", "Física", "Álgebra", "Dibujo"],
    "Calificación": [85, 78, 92, 88]
})

print("df1:")
print(df1)
print("\ndf2:")
print(df2)

df_concatenated = pd.concat([df1, df2], ignore_index=True)
print("\nDataFrame concatenado:")
print(df_concatenated)

df_merged = pd.merge(df1, df2, on='ID_Estudiante')
print("\nDataFrame merged:")
df_merged
```
