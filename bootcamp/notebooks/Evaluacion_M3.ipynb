```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('migracion.csv')
df.head()
```

```python
valores_perdidos = df.isnull().sum()
print("Valores perdidos por columna:\n", valores_perdidos)
```

```python
Q1_migrantes = df['Cantidad_Migrantes'].quantile(0.25)
Q3_migrantes = df['Cantidad_Migrantes'].quantile(0.75)
IQR_migrantes = Q3_migrantes - Q1_migrantes
print("Rango intercuartil Cantidad_Migrantes (IQR):\n", IQR_migrantes)

Q1_pib_or = df['PIB_Origen'].quantile(0.25)
Q3_pib_or = df['PIB_Origen'].quantile(0.75)
IQR_pib_or = Q3_pib_or - Q1_pib_or
print("Rango intercuartil PIB_Origen (IQR):\n", IQR_pib_or)

Q1_pib_dest = df['PIB_Destino'].quantile(0.25)
Q3_pib_dest = df['PIB_Destino'].quantile(0.75)
IQR_pib_dest = Q3_pib_dest - Q1_pib_dest
print("Rango intercuartil PIB_Destino (IQR):\n", IQR_pib_dest)

Q1_IDH_or = df['IDH_Origen'].quantile(0.25)
Q3_IDH_or = df['IDH_Origen'].quantile(0.75)
IQR_IDH_or = Q3_IDH_or - Q1_IDH_or
print("Rango intercuartil IDH_Origen (IQR):\n", IQR_IDH_or)

Q1_IDH_dest = df['IDH_Destino'].quantile(0.25)
Q3_IDH_dest = df['IDH_Destino'].quantile(0.75)
IQR_IDH_dest = Q3_IDH_dest - Q1_IDH_dest
print("Rango intercuartil IDH_Destino (IQR):\n", IQR_IDH_dest)

# Calcular los límites para la detección de outliers
lower_bound_migrantes = Q1_migrantes - 1.5 * IQR_migrantes
upper_bound_migrantes = Q3_migrantes + 1.5 * IQR_migrantes

lower_bound_pib_or = Q1_pib_or - 1.5 * IQR_pib_or
upper_bound_pib_or = Q3_pib_or + 1.5 * IQR_pib_or

lower_bound_pib_dest = Q1_pib_dest - 1.5 * IQR_pib_dest
upper_bound_pib_dest = Q3_pib_dest + 1.5 * IQR_pib_dest

lower_bound_idh_or = Q1_IDH_or - 1.5 * IQR_IDH_or
upper_bound_idh_or = Q3_IDH_or + 1.5 * IQR_IDH_or

lower_bound_idh_dest = Q1_IDH_dest - 1.5 * IQR_IDH_dest
upper_bound_idh_dest = Q3_IDH_dest + 1.5 * IQR_IDH_dest

# Filtrar el DataFrame para eliminar los outliers
df_filtered = df[
    (df['Cantidad_Migrantes'] >= lower_bound_migrantes) & (df['Cantidad_Migrantes'] <= upper_bound_migrantes) &
    (df['PIB_Origen'] >= lower_bound_pib_or) & (df['PIB_Origen'] <= upper_bound_pib_or) &
    (df['PIB_Destino'] >= lower_bound_pib_dest) & (df['PIB_Destino'] <= upper_bound_pib_dest) &
    (df['IDH_Origen'] >= lower_bound_idh_or) & (df['IDH_Origen'] <= upper_bound_idh_or) &
    (df['IDH_Destino'] >= lower_bound_idh_dest) & (df['IDH_Destino'] <= upper_bound_idh_dest)
]

print("\nDataFrame original:\n", df.shape)
print("\nDataFrame filtrado (sin outliers):\n", df_filtered.shape)
```

```python
mapeo_razones = {
  "Económica": "Trabajo",
  "Conflicto": "Guerra"
}

df['Razon_Migracion'] = df['Razon_Migracion'].replace(mapeo_razones)
df
```

```python
media_migrantes = df['Cantidad_Migrantes'].mean()
mediana_migrantes = df['Cantidad_Migrantes'].median()

print(f"Media de la cantidad de migrantes: {media_migrantes:.2f}")
print(f"Mediana de la cantidad de migrantes: {mediana_migrantes:.2f}")

pib_origen_promedio = df['PIB_Origen'].mean()
pib_destino_promedio = df['PIB_Destino'].mean()

print(f"PIB promedio de los países de origen: {pib_origen_promedio:.2f}")
print(f"PIB promedio de los países de destino: {pib_destino_promedio:.2f}")

conteo_razones = df['Razon_Migracion'].value_counts()
print("\nConteo de movimientos de migración por razón:")
conteo_razones
```

```python
df_grouped_reason = df.groupby('Razon_Migracion').agg(
    Total_Migrantes=('Cantidad_Migrantes', 'sum'),
    Media_IDH_Origen=('IDH_Origen', 'mean')
).reset_index()

df_grouped_reason_sorted = df_grouped_reason.sort_values(by='Total_Migrantes', ascending=False)

print("Agrupación por Razón de Migración y Media de IDH de Origen (ordenado por Total de Migrantes):")
df_grouped_reason_sorted
```
