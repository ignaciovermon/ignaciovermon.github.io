```python
class Libro():
  def __init__(self, titulo, autor, isbn):
    self.titulo = titulo
    self.autor = autor
    self.isbn = isbn

  def get_titulo(self):
    return self.titulo

  def get_autor(self):
    return self.autor

  def get_isbn(self):
    return self.isbn

  def descripcion(self):
    print(f'Titulo: {self.titulo}, Autor: {self.autor}, ISBN: {self.isbn}')

class Biblioteca():
  def __init__(self):
    self.libros = []

  def agregar_libro(self, libro):
    self.libros.append(libro)

  def mostrar_libros(self):
    for libro in self.libros:
      libro.descripcion()
```

```python
# Instanciar la clase Biblioteca
biblioteca = Biblioteca()

# Crear al menos dos libros
libro1 = Libro("Cien años de soledad", "Gabriel García Márquez", "978-3-16-148410-0")
libro2 = Libro("El amor en los tiempos del cólera", "Gabriel García Márquez", "978-3-16-148410-1")

# Añadir los libros a la biblioteca
biblioteca.agregar_libro(libro1)
biblioteca.agregar_libro(libro2)

# Mostrar los detalles de los libros almacenados
biblioteca.mostrar_libros()
```

```python
while True:
  try:
    print("Bienvenido a la Biblioteca")
    print("1. Agregar libro")
    print("2. Mostrar libros")
    print("3. Salir")

    opcion = int(input("Selecciona una opción: "))
    if opcion == 1:
      titulo = input("Ingresa el título del libro: ")
      autor = input("Ingresa el autor del libro: ")
      isbn = input("Ingresa el ISBN del libro: ")
      libro = Libro(titulo, autor, isbn)
      biblioteca.agregar_libro(libro)
      print("Libro agregado con éxito.")
    elif opcion == 2:
      print("Lista de libros:")
      biblioteca.mostrar_libros()
    elif opcion == 3:
      print("¡Hasta luego!")
      break

    else:
      print("Opción no válida. Por favor, selecciona una opción válida.")
  except ValueError:
    print("Por favor, ingresa un número válido.")
```
