```python
# prompt: Implementa un bucle for para mostrar el nombre y los comentarios de todos los estudiantes
# ingresados

estudiantes = []

while True:
  nombre_estudiante = str(input('Ingrese nombre del estudiante:'))
  nombre_materia1 = str(input('Ingrese nombre primera materia:'))
  nombre_materia2 = str(input('Ingrese nombre segunda materia:'))
  nombre_materia3 = str(input('Ingrese nombre tercera materia:'))
  materia1 = int(input('Ingrese nota primera materia:'))
  materia2 = int(input('Ingrese nota segunda materia:'))
  materia3 = int(input('Ingrese nota tercera materia:'))

  calificacion_estudiante = round((materia1 + materia2 + materia3)/3)

  comentario = ""
  if calificacion_estudiante >= 90 and calificacion_estudiante <= 100:
    comentario = "Esta aprobado. Excelente!"
  elif calificacion_estudiante >= 75 and calificacion_estudiante <89:
    comentario = "Esta aprobado. Bueno!"
  elif calificacion_estudiante >= 60 and calificacion_estudiante <75:
    comentario = "Esta aprobado, pero necesita mejorar!"
  elif calificacion_estudiante >= 0 and calificacion_estudiante < 60:
    comentario = "Esta reprobado. Necesita mejorar"
  else:
    comentario = "La calificacion ingresada no es valida"

  estudiantes.append({"nombre": nombre_estudiante, "comentario": comentario})
  print(f'El estudiante {nombre_estudiante} {comentario}')

  continuar = input("¿Desea ingresar otro estudiante? (s/n): ")
  if continuar.lower() != 's':
    break

print("\n--- Listado de Estudiantes ---")
for estudiante in estudiantes:
  print(f"Nombre: {estudiante['nombre']}, Comentario: {estudiante['comentario']}")
```

```python
# prompt: Usa una expresión ternaria para agregar una nota adicional en los comentarios si el
# estudiante tiene un promedio de 100 “¡Puntuación perfecta!

estudiantes = []

while True:
  nombre_estudiante = str(input('Ingrese nombre del estudiante:'))
  nombre_materia1 = str(input('Ingrese nombre primera materia:'))
  nombre_materia2 = str(input('Ingrese nombre segunda materia:'))
  nombre_materia3 = str(input('Ingrese nombre tercera materia:'))
  materia1 = int(input('Ingrese nota primera materia:'))
  materia2 = int(input('Ingrese nota segunda materia:'))
  materia3 = int(input('Ingrese nota tercera materia:'))

  calificacion_estudiante = round((materia1 + materia2 + materia3)/3)

  comentario = ""
  if calificacion_estudiante >= 90 and calificacion_estudiante <= 100:
    comentario = "Esta aprobado. Excelente!"
  elif calificacion_estudiante >= 75 and calificacion_estudiante <89:
    comentario = "Esta aprobado. Bueno!"
  elif calificacion_estudiante >= 60 and calificacion_estudiante <75:
    comentario = "Esta aprobado, pero necesita mejorar!"
  elif calificacion_estudiante >= 0 and calificacion_estudiante < 60:
    comentario = "Esta reprobado. Necesita mejorar"
  else:
    comentario = "La calificacion ingresada no es valida"

  # Usar una expresión ternaria para agregar la nota adicional si la calificación es 100
  comentario_final = comentario + (" ¡Puntuación perfecta!" if calificacion_estudiante == 100 else "")

  estudiantes.append({"nombre": nombre_estudiante, "comentario": comentario_final})
  print(f'El estudiante {nombre_estudiante} {comentario_final}')

  continuar = input("¿Desea ingresar otro estudiante? (s/n): ")
  if continuar.lower() != 's':
    break

print("\n--- Listado de Estudiantes ---")
for estudiante in estudiantes:
  print(f"Nombre: {estudiante['nombre']}, Comentario: {estudiante['comentario']}")
```
