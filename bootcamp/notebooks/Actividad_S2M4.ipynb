```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ejercicio.csv')
df.head()
```

```python
for column in df.columns:
    dtype = df[column].dtype
    unique_values = df[column].nunique()
    print(f"Columna: {column}, Tipo de dato: {dtype}")

    if dtype == 'object':
      print("  Tipo de variable: Categórica")
    elif dtype in ['int64', 'int32', 'int16']:
      print("  Tipo de variable: Cuantitativa Discreta")
    elif dtype in ['float64', 'float32']:
      print("  Tipo de variable: Cuantitativa Continua")
    else:
      print("  Tipo de variable: Desconocido")

    print("-" * 30)
```

```python
# Frecuencia para una variable categórica ('Género')
print("Tabla de frecuencia para variable categórica ('Género'):")
frecuencia_genero = df['Género'].value_counts()
print(frecuencia_genero)
print("-" * 30)

# Frecuencia para una variable cuantitativa discreta ('Edad')
print("Tabla de frecuencia para variable cuantitativa discreta ('Edad'):")
frecuencia_edad = df['Edad'].value_counts().sort_index()
print(frecuencia_edad)
print("-" * 30)
```
